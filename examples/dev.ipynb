```python
import architile
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection

all_tilings = architile.tiling.all_tilings
n = len(all_tilings)

fig, axes = plt.subplots(3, 4, figsize=(18, 12))
a = 0.5
axes = axes.flatten()
# np.unravel_index(8, axes.shape)
for i, tiling_cls in enumerate(all_tilings.values()):
    ax = axes[i]
    ax.set_aspect("equal")
    tiling = tiling_cls(a)

    nodes, edges = architile.tile_into_rectangle(
        (0, 0, 5, 5), tiling, theta=0.0, boundary="cut_fill"
    )

    lcol = LineCollection(nodes[edges], colors="black", alpha=0.8)
    ax.add_collection(lcol)
    ax.scatter(*nodes.T, s=10, alpha=0.5)

    ax.set_title(tiling_cls.__name__)
    ax.set_xticks([])
    ax.set_yticks([])
    # ax.margins(0)

axes[n].set_visible(False)

fig.tight_layout()
# fig.savefig("tilings.png", dpi=160)
```
